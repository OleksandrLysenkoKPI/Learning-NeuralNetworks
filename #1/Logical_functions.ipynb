```python
# Зважена сума S: S = x1*w1 + x2*w2 + x3*w3 + ... + xn * wn
# Функція активації F(S) вар'юється від випадку
```

```python
# Моделювання логічної функції "І" (AND)
def AND_func(data):
    print("{:<5} {:<5} {:<5} {:<5}".format("x1", "x2", "S", "Y"))
    print("-" * 20)
    
    for x1, x2 in data:
        w1, w2 = 1, 1
        S = x1*w1 + x2*w2
        
        FS = 1 if S >= 1.5 else 0
    
        print("{:<5} {:<5} {:<5} {:<5}".format(x1, x2, S, FS))

input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]
AND_func(input_data)
```

```python
# Моделювання логічної функції "АБО" (OR)
def OR_func(data):
    print("{:<5} {:<5} {:<5} {:<5}".format("x1", "x2", "S", "Y"))
    print("-" * 20)

    for x1, x2 in data:
        w1, w2 = 1, 1
        S = x1*w1 + x2*w2
        
        FS = 1 if S >= 0.5 else 0
    
        print("{:<5} {:<5} {:<5} {:<5}".format(x1, x2, S, FS))

input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]
OR_func(input_data)
```

```python
# Моделювання логічної функції "НІ" (NOT)
def NOT_func(data):
    print("{:<5} {:<5} {:<5}".format("x", "S", "Y"))
    print("-" * 14)

    for x in data:
        w = -1.5
        S = x*w
        
        FS = 1 if S >= -1 else 0
    
        print("{:<5} {:<5} {:<5}".format(x, S, FS))

input_data = [0, 1]
NOT_func(input_data)
```

```python
# Моделювання логічної функції "Виключне АБО" (XOR)
def XOR_func(data):
    print("{:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5}".format(
        "x1", "x2", "S1", "y1", "S2", "y2", "S3", "Y"
    ))
    print("-" * 45)

    # Ваги для 3-х нейронів
    w11, w12 =  1, -1  # Нейрон N1
    w21, w22 = -1,  1  # Нейрон N2
    w31, w32 =  1,  1  # Нейрон N3 (вихідний)

    # Порогова функція активації
    def step_activation(S):
        return 1 if S >= 0.5 else 0

    for x1, x2 in data:
        # Обчислення сум на вході кожного нейрона прихованого шару
        S1 = x1*w11 + x2*w12
        S2 = x1*w21 + x2*w22

        # Виходи нейронів прихованого шару
        y1 = step_activation(S1)
        y2 = step_activation(S2)

        # Сума на вході вихідного нейрона
        S3 = y1*w31 + y2*w32
        # Вихід XOR
        Y  = step_activation(S3)

        print("{:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5} {:<5}".format(
            x1, x2, round(S1, 2), y1, round(S2, 2), y2, round(S3, 2), Y
        ))

input_data = [(0,0), (0,1), (1,0), (1,1)]
XOR_func(input_data)
```
